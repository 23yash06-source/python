Question . 1.	Write a Python function to find the maximum and minimum numbers from a sequence of numbers

```python
def find_max_min(numbers):
    if len(numbers) == 0:
        return "Sequence is empty"

    maximum = numbers[0]
    minimum = numbers[0]

    for num in numbers:
        if num > maximum:
            maximum = num
        if num < minimum:
            minimum = num

    return maximum, minimum


nums = [10, 25, 3, 89, 45, 2]
max_num, min_num = find_max_min(nums)

print("Maximum number:", max_num)
print("Minimum number:", min_num)
```

Question . 2.	Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number.

```python
def sum_of_cubes(n):
    total = 0
    for i in range(1, n):
        total += i ** 3
    return total

num = int(input("Enter a positive integer: "))

if num > 0:
    result = sum_of_cubes(num)
    print("Sum of cubes of all positive integers smaller than", num, "is:", result)
else:
    print("Please enter a positive integer.")
```

Question . 3.	Write a Python function to print 1 to n using recursion.

```python
def print_numbers(n, current=1):
    if current > n:
        return
    print(current)
    print_numbers(n, current + 1)

n = int(input("Enter a number: "))
print_numbers(n)
```

Question. 4 Write a recursive function to print Fibonacci series upto n terms.

```python
def fibonacci(n, a=0, b=1, count=0):
    if count == n:
        return
    print(a, end=" ")
    fibonacci(n, b, a + b, count + 1)

n = int(input("Enter the number of terms: "))
fibonacci(n)
```

Question. 5 Write a lambda function to find volume of cone

```python
import math

cone_volume = lambda r, h: (1/3) * math.pi * r**2 * h

radius = float(input("Enter radius of cone: "))
height = float(input("Enter height of cone: "))

print("Volume of cone:", cone_volume(radius, height))
```

Question. 6 Write a lambda function which gives tuple of max and min from a list.

```python
max_min = lambda lst: (max(lst), min(lst))

numbers = [10, 6, 8, 90, 12, 56]
result = max_min(numbers)
print(result)
```

Question. 7 Write functions to explain the mentioned concepts:
a. Keyword argument
b. Default argument
c. Variable-length argument

```python
# Function using keyword arguments
def greet(name, age):
    print(f"Hello {name}, you are {age} years old.")

greet(age=25, name="Alice")
```

```python
# Function with default argument
def greet(name, age=18):
    print(f"Hello {name}, you are {age} years old.")

greet("Bob")
```

```python
# Function with variable-length arguments
def student_scores(name, *scores):
    print(f"Student: {name}")
    print(f"Scores: {scores}")
    print(f"Total: {sum(scores)}")

student_scores("Charlie", 80, 90, 75)
```

Question. 8 Write a program to check whether all the values in a dictionary are the same or not using
a lambda function.

```python
my_dict = {'a': 10, 'b': 10, 'c': 10}

all_same = lambda d: len(set(d.values())) == 1

result = all_same(my_dict)
print(result)
```
