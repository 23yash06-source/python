Question. Write a program to find simple interest.

```python
P = float(input("Enter the principal amount: "))
R = float(input("Enter the rate of interest: "))
T = float(input("Enter the time (in years): "))

SI = (P * R * T) / 100

print("Simple Interest:", SI)
```

Question. Write a program to find area of triangle when length of sides are given

```python
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))

s = (a + b + c) / 2

area = (s * (s - a) * (s - b) * (s - c)) ** 0.5

print("Area of the triangle:", area)
```

Question. Write a program to print truth table for bitwise operators (&, | and ^ operators)

```python
values = [0, 1]

print("A B | A&B A|B A^B")

for A in values:
    for B in values:
        print(A, B, "|", A & B, " ", A | B, " ", A ^ B)
```

Question. Write a program to find left shift and right shift values of a given number.

```python
num = int(input("Enter a number: "))
shift = int(input("Enter number of positions to shift: "))

left_shift = num << shift

right_shift = num >> shift

print("Original number:", num)
print("Left shift value:", left_shift)
print("Right shift value:", right_shift)
```

Question. Using membership operator find whether a given number is in sequence (10,20,56,78,89)

```python
sequence = (10, 20, 56, 78, 89)

num = int(input("Enter a number: "))

if num in sequence:
    print(num, "is present in the sequence")
else:
    print(num, "is not present in the sequence")
```
