Question. Find a factorial of given number.

```python
n = int(input("Enter a number: "))

fact = 1
for i in range(1, n + 1):
    fact *= i

print("Factorial of", n, "is", fact)
```

Question. 2Find whether the given number is Armstrong number

```python
n = int(input("Enter a number: "))

num_digits = len(str(n))
sum_power = 0
temp = n

while temp > 0:
    digit = temp % 10
    sum_power += digit ** num_digits
    temp //= 10

if sum_power == n:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

Question. 3Print Fibonacci series up to given term.

```python
n = int(input("Enter number of terms: "))

a, b = 0, 1

print("Fibonacci Series:")
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

Question. 4Write a program to find if given number is prime number or not

```python
n = int(input("Enter a number: "))

if n <= 1:
    print("Not a prime number")
else:
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            print("Not a prime number")
            break
    else:
        print("Prime number")
```

Question. 5Check whether given number is palindrome or not.

```python
n = int(input("Enter a number: "))

temp = n
rev = 0

while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp //= 10

if rev == n:
    print("Palindrome number")
else:
    print("Not a palindrome number")
```

Question. 6Write a program to print sum of digits.


```python
n = int(input("Enter a number: "))

sum_digits = 0

while n > 0:
    digit = n % 10
    sum_digits += digit
    n //= 10

print("Sum of digits:", sum_digits)
```

Question. 7Count and print all numbers divisible by 5 or 7 between 1 to 100.

```python
count = 0

print("Numbers divisible by 5 or 7 between 1 and 100:")

for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        print(i, end=" ")
        count += 1

print("\n\nTotal count:", count)
```

Question. 10 Write a program to print the following pattern
123454321
1234 *4321
123 * * 321
12 * * * 21
1 * * * * 1

```python
n = 5

for i in range(n):
    for j in range(1, n - i):
        print(j, end="")

    for j in range(i):
        print(" *", end="")

    if i > 0:
        print(" ", end="")


    for j in range(n - i - 1, 0, -1):
        print(j, end="")

    print()
```

Question. 11Write a program to print the sum of the following series
1+ ½ + 1/3 + ¼ +….+1/n

```python
n = int(input("Enter the value of n: "))

sum_series = 0.0

for i in range(1, n + 1):
    sum_series += 1 / i

print("Sum of the series is:", sum_series)
```
